```python
from setup_dask import start_dask
from preprocess import load_and_clean_data
from performance_compare import benchmark_sklearn, benchmark_dask
import pandas as pd
import matplotlib.pyplot as plt

# Start Dask
client = start_dask()

# Load and preprocess
df = load_and_clean_data()

# Run both benchmarks
results = []
results.append(benchmark_sklearn(df))
results.append(benchmark_dask(df))

# Visualization
df_results = pd.DataFrame(results)

# Bar chart of accuracy
df_results.set_index("framework")["accuracy"].plot(kind="bar", title="Accuracy Comparison", ylim=(0, 1), ylabel="Accuracy", color="green")
plt.show()

# Bar chart of training time
df_results.set_index("framework")["time"].plot(kind="bar", title="Training Time Comparison", ylabel="Seconds", color="orange")
plt.show()
```
